## Pie Chart (3 points)

```python
# Create a pie chart from raw values

import matplotlib.pyplot as plt

slices = [56, 66, 24, 124, 47]
labels = ["US", "UK", "MX", "CA", "JP"]

plt.pie(slices, labels=labels, autopct="%1.1f%%", startangle=90)

plt.title("Distribution Across Countries")

plt.show()
```

## Histogram (3 points)

```python
# Create a histogram from random data

import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)

plt.hist(n, bins=30, color='yellow', edgecolor='black')

plt.title('Histogram of Random Data')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

## Scatterplot with Trendline (3 points)

```python
# Create a scatterplot with a trendline from two lists of values
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

sns.set(style="whitegrid")

num_points = 50

x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

# If you want them combined into a dataframe
df = pd.DataFrame({'x': x, 'y': y})

sns.regplot(x='x', y='y', data=df, ci=None, line_kws={'color': 'red'})

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatterplot with Trendline')

plt.show()
```
